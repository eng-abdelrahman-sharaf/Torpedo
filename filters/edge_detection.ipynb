```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("images/Sudoku.png")
# image = cv.resize(image, None , None , fx = 5 , fy = 5 , interpolation = cv.INTER_CUBIC)
src = cv.GaussianBlur(image, (3, 3), 0)
src_gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
ksize = 5
```

```python
plt.figure(figsize=(20,20))

plt.subplot(221)
plt.imshow(image,cmap="gray")
plt.title('original')
plt.axis('off')

plt.subplot(222)
edged = cv.Laplacian(src_gray, -1, ksize=ksize)
plt.imshow(edged,cmap="gray")
plt.title('laplacian')
plt.axis('off')

Sobel_X=cv.Sobel(src_gray,-1,1,0) 
Sobel_Y=cv.Sobel(src_gray,-1,0,1)
Sobel_or=cv.bitwise_or(Sobel_X,Sobel_Y)
plt.subplot(223)
plt.imshow(Sobel_or,cmap="gray")
plt.title("Sobel")
plt.axis("off")

canny=cv.Canny(src_gray,22,80)
plt.subplot(224)
plt.imshow(canny,cmap="gray")
plt.title("Canny")
plt.axis("off")

plt.show()
```
